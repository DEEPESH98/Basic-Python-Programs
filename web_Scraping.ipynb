```
from bs4 import BeautifulSoup
```

```
import requests
```

```
url = 'https://en.wikipedia.org/wiki/Iris_(anatomy)'
```

```
# we are sending the request to the url and taking the response 
response=requests.get(url)
```

```
# <Response [200]> // re response code he jese http 80 pe chaleta he
response 
```

```
print(response)
```

```
response.text
```

```
# parshing Library - html.parser and second is lxml
soup=BeautifulSoup(response.text, "lxml")
```

```
soup
```

```
for Alink in soup.find_all('a'): # ya soup bhii chale ga
  print(Alink)
  
```

```
# for abherf in Alink.find_all('href')
for Alink in soup.find_all('a'): # ya soup b hii chale ga
  print(Alink.get('href'))
  
```

```
for ptag in soup.find_all('p'): 
  print(ptag)
  # now second loop for b tag
  for btag in ptag.find_all('b'):
    print(btag)
```
